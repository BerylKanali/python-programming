```python
##create a table called email with two columns, email and count,showing all emails that start with from how many times they occur 
```

```python
import sqlite3
sqlite3.version
```

```python
con=sqlite3.connect('C:\\Users\\USER\\student.db')
```

```python
cur=con.cursor()
```

```python
fhand=open("mbox.txt")
email=dict()
for line in fhand:
    line=line.rstrip()
    if not line.startswith("From "):
        continue
    words=line.split()
    print(words[1])
    if words[1] not in email:
        email[words[1]]=1
    else:
        email[words[1]] = email[words[1]] +1
```

```python
email
```

```python
for key,value in email.items():
    cur.execute('insert into emails (email,count) values (?,?)',(key,value))
```

```python
con.commit()
```
