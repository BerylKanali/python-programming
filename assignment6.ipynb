### most frequent day

```python
fhand = open('mbox.txt') 
for line in fhand:     
    line = line.rstrip()    
    if not line.startswith('From ') : 
        continue    
    words = line.split()    
    print (words[2])
```

```python
fhand = open('mbox.txt') 
DAY=dict()
for line in fhand:     
    line = line.rstrip()    
    if not line.startswith('From ') : 
        continue    
    words = line.split()    
    print (words[2])
    if words[2] not in DAY:
        DAY[words[2]]=1
    else:
        DAY[words[2]]=DAY[words[2]]+1
```

```python
DAY
```

```python
day=list()
for a,b in DAY.items():
    day.append( (b,a) )
print(day)
```

```python
day.sort(reverse=True)
print(day)
print("most frequent day is", day[0])
```

### most frequent day of the month

```python
fhand = open('mbox.txt') 
frequent=dict()
for line in fhand:     
    line = line.rstrip()    
    if not line.startswith('From ') : 
        continue    
    words = line.split() 
    print(words[4])
    if words[4] not in frequent:
        frequent[words[4]]=1
    else:
        frequent[words[4]]=frequent[words[4]]+1
        
```

```python
frequent
```

```python
date=list()
for a,b in frequent.items():
    date.append( (b,a) )
print(date)
```

```python
date.sort(reverse=True)
print(date)
print("most frequent day of the month is", date[0])
```
